```python
import re 
import math

vocabulario = []
num_docs = 0 #numero de documentos
tf = dict() # frecuencia tf {<palabra>: <[freceuncia D1,freceuncia D2, ... freceuncia Dn]>}
df = dict() # documentos que contienen la palabra {<palabra>:<cantidad documentos>}
idf = dict() #log(N/df) {<palabra>: <idf>}
tfidf = dict() #tf x idf {<palabra>: <[W_i1, W_i2, ... W_ij]>}
```

```python
def count_words(text, num):
    """Cuenta cuantas veces una palabra occure en el texto"""
    counts = dict()  # diccionario de pares { <word>: <count> } 
 
    # Dividir el texto en tokens (palabras), quitando la puntuacion.
    # Usar expresiones regulares para dividr segun caracteres no-alfanumericos '\w'
    matchObjs = re.findall(r'[\w]+', text)
    
    #print(matchObjs)
    # Conteo usando el diccionario.
    for obj in matchObjs:
        if obj in counts:
            counts[obj] += 1 #Adiciona 1 a una entrada existente
            calcular_tf(obj, num)
        else:
            counts[obj] = 1 #Crea un nuevo indice/palabra en el diccionario.
            if obj not in vocabulario:
                    vocabulario.append(obj) #agarega al vocabilario
            calcular_tf(obj, num)
            if obj in df:
                df[obj]+=1
            else:
                df[obj]=1
    for word, count in tf.items():
        aux=tf.get(word)
        aux[num]=aux[num]/len(matchObjs)
        tf[word]=aux
    
    return counts
```

```python
def calcular_idf():
    for palabra, n in df.items():
        idf[palabra] = math.log(num_docs/n)
```

```python
def calcular_tf(obj, num):
    if obj in tf:
        aux=tf.get(obj)
        aux[num]+=1
        tf[obj]=aux
    else:
        aux=[0]*num_docs
        aux[num]+=1
        tf[obj]=aux
```

```python
def calcular_tfidf():
    for palabra, frecuencia in tf.items():
        aux = []
        for f in frecuencia:
            aux.append(f*idf.get(palabra))
        tfidf[palabra]=aux
        
```

```python
def extraer_doc(text):
    global num_docs
    print("Documentos")
    # Conversion a minuscula
    text2 = text.lower()
     #separar por lineas
    docs = re.split('\n|\r', text2)
    num_docs=len(docs)
    for d in range(num_docs):
        count_words(docs[d],d)
    calcular_idf()
    calcular_tfidf()
    return docs
```

```python
def matrix (text):
    extraer_doc(text)
    print_matriz()
```

```python
def print_matriz():
    aux= "{0:20s}".format("Documentos")
    for n in range(num_docs):
        aux+= "{0:10s}".format("D"+str(n+1))
    for palabra, wij in tfidf.items():
        aux+="\n"
        aux+= "{0:15}".format(palabra)
        for f in wij:
            aux+= "{0:10f}".format(f)
    print(aux)
    with open("matrix1.txt", "w") as archivo:
        archivo.writelines([aux])
```
